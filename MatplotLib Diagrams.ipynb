```python
# importing the required module 
import matplotlib.pyplot as plt 

# x axis values 
x1 = [2,6,10,20,30,40,50,60,70] 
# corresponding y axis values 
y1 = [2.7473,2.5821,2.5045,2.455,2.4212,2.4015,2.3991,2.4284,2.5787]

# plotting the points  
plt.plot(x1, y1, label = "Weighted")

# line 2 points 
x2 = [2,6,10,20,30,40,50,60,70]  
y2 = [2.7446,2.5776,2.4994,2.4479,2.4111,2.3866,2.3686,2.2549,2.2562] 
# plotting the line 2 points  
plt.plot(x2, y2, label = "Normal") 
  
# naming the x axis 
plt.xlabel('K') 
# naming the y axis 
plt.ylabel('Mean Absolute Error')

# show a legend on the plot 
plt.legend() 

# giving a title to my graph 
plt.title('Adjusted Cosine Similarity') 

# function to show the plot 
plt.show()
```
